```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
with open('./cnn_training_results.pkl', 'rb') as f:
    cnn_train_losses, cnn_eval_losses, cnn_train_corrects, cnn_eval_corrects, cnn_best_accuracy = pickle.load(f)

cnn_train_accuracies = [correct / int(49479) for correct in cnn_train_corrects]
cnn_eval_accuracies = [correct / int(11974) for correct in cnn_eval_corrects]
```

```python
plt.figure(figsize=(10,6))
plt.plot(cnn_train_losses, label="Training Loss", color='blue')
plt.plot(cnn_eval_losses, label="Validation Loss", color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title("Training and Validation Loss over Epochs (CNN model)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(cnn_train_accuracies, label='Training Accuracy', marker='o', color='blue')
plt.plot(cnn_eval_accuracies, label='Validation Accuracy', marker='o', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy Over Epochs (CNN model)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
Lr_results = pd.read_csv('./LR_results.csv')
simple_lr_acc = Lr_results.iloc[0,0] * 100
crva_lr_acc = Lr_results.iloc[1,0] * 100
```

```python
accuracy_data = pd.DataFrame({'Model': ['Logistic Regression Without Cross Validation', 'Logistic Regression With Cross Validation', 'CNN'], 'Accuracy': [simple_lr_acc,crva_lr_acc,cnn_best_accuracy]})
sns.barplot(y='Model', x='Accuracy', data=accuracy_data, hue='Model')
plt.title('Model Accuracies', fontsize=14)
```
